##Construya expresiones regulares para los siguientes tokens:

1.   Identificadores del lenguaje BASH
2.   Cadenas python multi línea de python
3.   Correo electrónico
4.   URL'S



```python
#EXPRESIÓN REGULAR PARA IDENTIFICAR COMANDOS BASH
p = re.compile('([a-z]*)+(\s\W(\s)?\w*)?')
```

```python
list_expres = ['dir', 'rm -d', 'route -n']

p.match('rem -d')
```

```python
#CADENAS PYTHON MULTI LÍNEA DE PYTHON
p2 = re.compile('^(\W*(\S\s)*?)[(\w*(\S\s)*?)]*(\W*)$')
```

```python
p2.match("""python
ada
asda """)
```

```python
#CORREOS ELECTRÓNICOS
p3 = re.compile('(\w*(\.*)?)*\@\w*\.\w*')
```

```python
p3.match("mark08.10.2001@gmail.com")
```

```python
#URL'S
p4 = re.compile('(http|https)://[a-zA-Z0-9.-]+\.[a-zA-Z]{2,}(?::\d+)?(/[^ ]*)?$')
```

```python
p4.match('https://www.youtube.com/watch?v=z1fadkdxAX0')
```
